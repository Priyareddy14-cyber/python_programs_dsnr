````python
"""
write a program to find the grades based on student marks
1. if marks in b/w 80 to 100 grade A
2. if marks in b/w 60 to 79 grade B
3. if marks in b/w 40 to 59 grade C
4. if marks in b/w 30 to 39 grade D
5. if marks in b/w 0 to 24 grade FAIL 
6. if marks other than above output should be please check valid marks
"""

marks=int(input("please enter student marks here:"))
if(marks>=80 and marks<=100):
    print("A GRADE")
elif(marks>=60 and marks<=79):
    print("B GRADE")
elif(marks>=40 and marks<=59):
     print("C GRADE")
elif(marks>=30 and marks<=39):
    print("D GRADE")
elif(marks>=0 and marks<=24):
    print("FAIL")
else:
    print("INVALID")
   

````

````python
n1=int(input("please enter 1st number here:"))
n2=int(input("please enter 2nd number here:"))
n3=int(input("please enter 3rd number here:"))
if(n1>=n2 and n1>=n3):
    print(n1,"is largest number")
elif(n2>=n1 and n2>=n3):
    print(n2,"is largest number")
else:
    print(n3,"is largest number")
    
````
