````python
#write a program to iterate 1 to 10 numbers
for i in range(1,10):
    print(i)
````

````python
# write a program to print names using for loop
names={"Priya","nakshatra","kundana","jetavya"}
for n in names:
    print("name of my friend is",n)
````

````python
#write a program to print even numbers in between 1 to 20 using for loop
for i in range(1,21):
    if(i%2==0):
        print(i,end=",")
````

````python
#write a program to print odd numbers in between 50 to 100 using for loop
for i in range(50,101):
    if(i%2!=0):
        print(i,end=",")
````

````python
#write a program to find the vowels from string using for loop
name=str(input("please enter your name here:"))
print(name)
for char in name:
    if(char in ["a","e","i","o","u"]):
        print(char)
````

````python
#write a program to find the non vowels from string using for loop
name=str(input("please enter your name here:"))
print(name)
for char in name:
    if(char  not in ["a","e","i","o","u"]):
        print(char,end=">")
````

````python
#write a program to find the sum of all numbers in between 1 to 20 using for loop
total = 0

# Loop through numbers from 1 to 20
for i in range(1, 21):
    total += i

# Display the result
print("The sum of numbers from 1 to 20 is:", total)
````

````python
#write a program to find the sum of all odd numbers in between 1 to 1000 using for loop
total = 0

# Loop through numbers from 1 to 1000
for i in range(1, 1001):
    if(i%2!=0):
        total += i
        

# Display the result
print("The sum of all odd numbers from 1 to 1001 is:", total)
````

````python
#write a program to find the product of all even numbers in between 500 to 600 using for loop
total = 1

# Loop through numbers from 500 to 600
for i in range(500, 601):
    if(i%2==0):
        total *= i
        

# Display the result
print("The product of all odd numbers from 500 to 600 is:", total)
````

````python
#write a program to find the numbers in between 1to 20 using while loop
sn=1
en=20
while(sn<=en):
    print(sn)
    sn=sn+1
````
